## Nuevos dataframes

Para facilitar los analisis, hay que limpiar y transformar los datos de la base de datos. Enfasis en el transformar, pues para hacer de manera mas sencilla los graficos, lo ideal es usar dataframes que surgan de hacer joins entre las distintas tablas. Es de estos nuevos dataframes que nosotros haremos nuestros analisis.

A continuacion planteamos los dataframes a crear:

- df_player_latest: Una fila por jugador, usando el dato mas reciente de Player_Attributes para cada player_api_id, unido con Player. Este dataset sirve para tener acceso a los nombres, edades, altura, peso y atributos de cada jugador en un mismo dataframe.
- df_team_match: una fila por equipo por partido. Cada fila sale de Match, separando local y visitante. Debe incluir team_api_id, opponent_api_id, season, league_id, country_id, goals_for, goals_against, diferencia de gol, puntos [3,1,0], condición local/visitante y formación inferida.
- df_team_season: una fila por equipo y temporada, agregando desde df_team_match puntos por partido, diferencia de gol por partido, win rate, goles a favor y en contra, y uniéndolo con un promedio por temporada de Team_Attributes.

Voy a hacer esto en [esta](Libreta_procesamiento_de_datos.ipynb) libreta de jupyter. **Por favor leanla**, ya que ahi se van a mostrar los dataframes con los cuales van a generar todas sus graficas (De hecho, pueden descargar esa libreta y ahi editar para generar el codigo de sus partes!, ya despues yo paso eso al script final).

### Scouting y atributos de jugadores

Los analisis planteados en esta categoria se centran en analizar los atributos de jugadores para mejorar la toma de desiciones a la hora de realizar scouting. El objetivo entonces era proveer informacion util para que el director tecnico tenga una mejor idea de como se relacionan los atributos, que atributos buscar en un jugador, etc.

#### Idea 01: Heatmap de correlacion entre atributos de jugadores

- [x] Realizada?

- Tema: Que atributos de jugador aportan informacion nueva y cuales son redundantes.

- Que se calcula?: Una matriz de correlacion entre variables numericas de Player_Attributes, idealmente usando correlacion de Spearman y separando jugadores de campo y porteros.

### Formaciones y tacticas

#### Idea 11: Boxplot de rendimiento por formacion

- Tema: Rendimiento deportivo segun formacion tactica.
- Pregunta: Que formaciones parecen producir mejores resultados por partido?
- Que se calcula?: La distribucion de puntos por partido o diferencia de gol para cada formacion inferida.
- Datos a usar: df_team_match
- Grafico: Boxplot por formacion.
- Precauciones: Exigir un minimo de partidos por formacion; no interpretar esto como causalidad directa; si pueden, revisar por separado local y visitante o controlar esa diferencia; excluir partidos sin formacion inferida confiable.
- Como identificar si el resultado es relevante?: Si algunas formaciones tienen medianas mas altas, menor dispersion o ambos, con tamano de muestra suficiente.

#### Idea 12: Grafico de barras apiladas de victoria/empate/derrota por formacion

- Como identificar si el resultado es relevante?: Si algunas formaciones muestran una proporcion visiblemente mayor de victorias o una proporcion menor de derrotas de forma consistente.
- Insight esperado: Identificar esquemas que parecen mas seguros y otros que parecen mas arriesgados o inestables.
